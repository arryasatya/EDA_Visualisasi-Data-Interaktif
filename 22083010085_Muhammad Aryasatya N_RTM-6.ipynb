```python
import pandas as pd

data_iris = pd.read_csv("iris.csv")
data_iris
```

```python
import dash

#mengimpor modul dash untuk membuat elemen html dan kompenen interaktif
from dash import dcc, html

#mengimpor kelas input dan output dari callback
from dash.dependencies import Input, Output

#mengimpor library plotly express untuk visualisasi data
import plotly.express as px

#membuat objek dash sebagai dasar aplikasi/tampilan visualisasi
app = dash.Dash(__name__)

#membuat layout/tata letak dari tampilan visualisasi
app.layout = html.Div([
     
    #tata letak untuk variabel "X"
    html.Label("Variabel X : "), dcc.Dropdown(id = "dropdown-x",options = [
            {"label" : "Sepal Length (Cm)", "value" : "sepal_length"},
            {"label" : "Sepal Width (Cm)", "value" : "sepal_width"},
            {"label" : "Petal Length (Cm)", "value" : "petal_length"},
            {"label" : "Petal Width (Cm)", "value" : "petal_width"}], 
                                              value = "sepal_length"),
    
    #tata letak untuk variabel "Y"
    html.Label("Variabel Y : "), dcc.Dropdown(id = "dropdown-y", options=[
            {"label" : "Sepal Length (Cm)", "value" : "sepal_length"},
            {"label" : "Sepal Width (Cm)", "value" : "sepal_width"},
            {"label" : "Petal Length (Cm)", "value" : "petal_length"},
            {"label": "Petal Width", "value": "petal_width"}],
                                              value = "sepal_width"),
    
    #penampilan scatter plot
    dcc.Graph(id = "scatter-plot"),
      
    #tata letak untuk "jumlah kelas"
    html.Label("Jumlah Kelas Iris : "),
    dcc.Dropdown(id = "dropdown-class",options=[
            {"label" : "3 Kelas", "value" : 3},
            {"label" : "2 Kelas", "value" : 2},
            {"label" : "1 Kelas", "value" : 1}],
                 value=3)])

#membuat callback untuk mengupdate visualisasi scatter plot sesuai inputan
@app.callback(
    Output("scatter-plot", "figure"),
    [Input("dropdown-x", "value"),
     Input("dropdown-y", "value"),
     Input("dropdown-class", "value")])

#define function untuk update visualisasi scatter plot
def update_scatter_plot(selected_x, selected_y, selected_class):
    filtered_data = data_iris[data_iris["species"].isin(data_iris["species"].unique()[:selected_class])]
    
    fig = px.scatter(filtered_data, x = selected_x, y = selected_y, color = "species", title=f"Scatter Plot Dataset Iris {selected_x} VS {selected_y}",
        labels = {"species" : "Species"})

    return fig

#menjalankan visualisasi yang telah dibuat
if __name__ == "__main__":
    app.run_server(debug = True)
```
